```python
def plot_image(image, title):
  """
  Esta función recibe una imagen y la muestra en pantalla ajustando
  el tamaño en base a las dimensiones del array de la imagen.
  Entrada: Imagen como array
  Salida: Imagen en pantalla
  """
  # plt.figure(figsize=(image.shape[0]*2/100, image.shape[1]*2/100))
  plt.imshow(image)
  plt.title(title)
  plt.axis('off')
  plt.show()
```

```python
def create_subplots(data_list, titles):
  """
  Esta función recibe un conjunto de datos (imágenes) para imprimir en
  pantalla junto con sus respectivos títulos.
  Entrada: Lista de imágenes y lista de títulos
  Salida: Imágenes en pantalla
  """
  num_images = len(data_list)
  num_cols = 3
  num_rows = (num_images + num_cols - 1) // num_cols
  plt.figure(figsize=(16, 12))
  for i in range(num_images):
    plt.subplot(num_rows, num_cols, i + 1)
    plt.imshow(data_list[i])
    plt.title(titles[i])
    plt.axis('off')  # Oculta los ejes
    plt.tight_layout()
  plt.show()
```

Construir una función debidamente documentada, que reciba un parámetro llamado "color", con el cual se indique los colores de caramelos que se quieren mostrar en la siguiente imagen:

```python
#@title Función
class color_ID:
  def __init__(self, imagen):
    """
    Inicializa una clase para guardar la
    imagen con que se quiere trabajar
    Entrada: Imagen a procesar
    """
    self.imagen = imagen
    self.hsv = cv2.cvtColor(imagen, cv2.COLOR_RGB2HSV)  #Conversión a HSV
    self.kernel = np.ones((5,5), np.uint8)  #Kernel para pasar por las imágenes

  def amarillo(self):
    """
    Identificación del color amarillo
    Entrada: None
    Salida: Imagen con objetos de color amarillo
    """
    #Creación de las máscaras
    low_yellow = np.array([20,100,100])
    upper_yellow = np.array([30,255,255])
    #Aplicación de la máscara
    mask = cv2.inRange(self.hsv, low_yellow, upper_yellow)
    res = cv2.bitwise_and(self.imagen, self.imagen, mask=mask)
    # Aplica una operación de opening para eliminar el ruido tipo salt
    op = cv2.morphologyEx(res, cv2.MORPH_OPEN, self.kernel)
    # Aplica una operación de closing después de la opening para eliminar tanto el ruido tipo salt como el ruido tipo pepper
    cl = cv2.morphologyEx(op, cv2.MORPH_CLOSE, self.kernel)
    return cl

  def verde(self):
    """
    Identificación del color verde
    Entrada: None
    Salida: Imagen con objetos de color verde
    """
    #Creación de las máscaras
    low_green = np.array([40,100,100])
    upper_green = np.array([80,255,255])
    #Aplicación de la máscara
    mask = cv2.inRange(self.hsv, low_green, upper_green)
    res = cv2.bitwise_and(self.imagen, self.imagen, mask=mask)
    # Aplica una operación de opening para eliminar el ruido tipo salt
    op = cv2.morphologyEx(res, cv2.MORPH_OPEN, self.kernel)
    # Aplica una operación de closing después de la opening para eliminar tanto el ruido tipo salt como el ruido tipo pepper
    cl = cv2.morphologyEx(op, cv2.MORPH_CLOSE, self.kernel)
    return cl

  def azul(self):
    """
    Identificación del color azul
    Entrada: None
    Salida: Imagen con objetos de color azul
    """
    #Creación de las máscaras
    low_blue = np.array([100,100,100])
    upper_blue = np.array([140,255,255])
    #Aplicación de la máscara
    mask = cv2.inRange(self.hsv, low_blue, upper_blue)
    res = cv2.bitwise_and(self.imagen, self.imagen, mask=mask)
    # Aplica una operación de opening para eliminar el ruido tipo salt
    op = cv2.morphologyEx(res, cv2.MORPH_OPEN, self.kernel)
    # Aplica una operación de closing después de la opening para eliminar tanto el ruido tipo salt como el ruido tipo pepper
    cl = cv2.morphologyEx(op, cv2.MORPH_CLOSE, self.kernel)
    return cl

  def purpura(self):
    """
    Identificación del color azul
    Entrada: None
    Salida: Imagen con objetos de color azul
    """
    #Creación de las máscaras
    low_purple = np.array([120,50,50])
    upper_purple = np.array([160,255,255])
    #Aplicación de la máscara
    mask = cv2.inRange(self.hsv, low_purple, upper_purple)
    res = cv2.bitwise_and(self.imagen, self.imagen, mask=mask)
    # Aplica una operación de opening para eliminar el ruido tipo salt
    op = cv2.morphologyEx(res, cv2.MORPH_OPEN, self.kernel)
    # Aplica una operación de closing después de la opening para eliminar tanto el ruido tipo salt como el ruido tipo pepper
    cl = cv2.morphologyEx(op, cv2.MORPH_CLOSE, self.kernel)
    return cl

  def rosado(self):
    """
    Identificación del color rosado
    Entrada: None
    Salida: Imagen con objetos de color rosado
    """
    #Creación de las máscaras
    low_pink = np.array([140,50,50])
    upper_pink = np.array([170,200,255])
    #Aplicación de la máscara
    mask = cv2.inRange(self.hsv, low_pink, upper_pink)
    res = cv2.bitwise_and(self.imagen, self.imagen, mask=mask)
    # Crea un kernel para operaciones morfológicas
    kernel = np.ones((6,6), np.uint8)
    # Aplica una operación de opening para eliminar el ruido tipo salt
    op = cv2.morphologyEx(res, cv2.MORPH_OPEN, kernel)
    # Aplica una operación de closing después de la opening para eliminar tanto el ruido
    cl = cv2.morphologyEx(op, cv2.MORPH_CLOSE, kernel)
    return cl

  def rojo(self):
    """
    Identificación del color naranja
    Entrada: None
    Salida: Imagen con objetos de color naranja
    """
    # Creación de la máscara
    low_red = np.array([170,50,50])
    upper_red = np.array([180,255,255])
    #Aplicación de la máscara
    mask = cv2.inRange(self.hsv, low_red, upper_red)
    res = cv2.bitwise_and(self.imagen, self.imagen, mask=mask)
    # Aplica una operación de opening para eliminar el ruido tipo salt
    op = cv2.morphologyEx(res, cv2.MORPH_OPEN, self.kernel)
    # Aplica una operación de closing después de la opening para eliminar tanto el ruido
    cl = cv2.morphologyEx(op, cv2.MORPH_CLOSE, self.kernel)
    return cl


  def naranja(self):
    """
    Identificación del color naranja
    Entrada: None
    Salida: Imagen con objetos de color naranja
    """
    # Creación de la máscara
    low_orange = np.array([0,100,100])
    upper_orange = np.array([10,255,255])
    #Aplicación de la máscara
    mask = cv2.inRange(self.hsv, low_orange, upper_orange)
    res = cv2.bitwise_and(self.imagen, self.imagen, mask=mask)
    # Aplica una operación de opening para eliminar el ruido tipo salt
    op = cv2.morphologyEx(res, cv2.MORPH_OPEN, self.kernel)
    # Aplica una operación de closing después de la opening para eliminar tanto el ruido
    cl = cv2.morphologyEx(op, cv2.MORPH_CLOSE, self.kernel)
    return cl

  def negro(self):
    """
    Identificación del color negro
    Entrada: None
    Salida: Imagen con objetos de color negro
    """
    # Creación de la máscara
    low_black = np.array([0,0,0])
    upper_black = np.array([180,255,50])
    #Aplicación de la máscara
    mask = cv2.inRange(self.hsv, low_black, upper_black)
    res = cv2.bitwise_and(self.imagen, self.imagen, mask=mask)
    # Aplica una operación de opening para eliminar el ruido tipo salt
    op = cv2.morphologyEx(res, cv2.MORPH_OPEN, self.kernel)
    # Aplica una operación de closing después de la opening para eliminar tanto el ruido
    cl = cv2.morphologyEx(op, cv2.MORPH_CLOSE, self.kernel)
    return cl

  def cafe(self):
    """
    Identificación del color cafe
    Entrada: None
    Salida: Imagen con objetos de color cafe
    """
    # Creación de la máscara
    low_brown = np.array([0,20,0])
    upper_brown = np.array([30,100,50])
    #Aplicación de la máscara
    mask = cv2.inRange(self.hsv, low_brown, upper_brown)
    res = cv2.bitwise_and(self.imagen, self.imagen, mask=mask)
    # Aplica una operación de opening para eliminar el ruido tipo salt
    op = cv2.morphologyEx(res, cv2.MORPH_OPEN, self.kernel)
    # Aplica una operación de closing después de la opening para eliminar tanto el ruido
    cl = cv2.morphologyEx(op, cv2.MORPH_CLOSE, self.kernel)
    return cl
```

# PUNTO #2
Construir una función debidamente documentada, que reciba un parámetro llamado "Operacion_Morfologica", el cual permitirá indicar a cual de los siguientes cuatro estados se quiere llevar la imagen "original". Lo anterior aplicando las cuatro operaciones morfólogicas vistas en clase a pedal, use cualquier imagen.

```python
def operacion(imagen, operacion_morfologica=""):
  """
  Esta función recibe una imagen y le aplica una operación morfológica dada
  (erosion, dilatacion, apertura, cierre)

  Entrada: Imagen a procesar y operación morfológica a aplicar
  Salida: Imagen con operación morfológica aplicada
  """
  kernel = np.ones((5,5), np.uint8)
  rows, cols, channels = imagen.shape
  struct_rows, struct_cols = kernel.shape
  struct_center_row, struct_center_col = struct_rows // 2, struct_cols // 2
  result = imagen.copy()
  """
  Para la erosión y dilatación tomamos el kernel planteado y recorremos la imagen
  comparando los pixeles para determinar si se deja o no el valor de la imagen.
  """
  if operacion_morfologica == "erosion":
    for i in range (struct_center_row, rows - struct_center_row):
      for j in range (struct_center_col, cols - struct_center_col):
        #roi es la Región de Interés que se utiliza para comparar con el kernel
        #y la imagen original
        roi = imagen[i - struct_center_row: i + struct_center_row + 1,
                    j - struct_center_col: j + struct_center_col + 1,]
        for c in range(channels):
          #Realizamos el proceso para cada máscara RGB
          if np.all(roi[:, :, c] >= kernel):
            result[i, j, c] = np.min(roi[:, :, c])
          else:
            result[i, j, c] = 0

  if operacion_morfologica == 'dilatacion':
    for i in range (struct_center_row, rows - struct_center_row):
      for j in range (struct_center_col, cols - struct_center_col):
        roi = imagen[i - struct_center_row: i + struct_center_row + 1,
                    j - struct_center_col: j + struct_center_col + 1,]
        for c in range(channels):
          if np.any(roi[:, :, c] * kernel):
            result[i, j, c] = np.max(roi[:, :, c])
          else:
            result[i, j, c] = 0

  """
  Para los siguientes procesos repetimos las funciones de erosión y dilatación
  invirtiendo el orden de las operaciones.
  """
  if operacion_morfologica == 'apertura':
    OP = operacion(imagen, 'erosion')
    result = operacion(OP, 'dilatacion')

  if operacion_morfologica == 'cierre':
    CL = operacion(imagen, 'dilatacion')
    result = operacion(CL, 'erosion')

  return result
```

```python
data = [cafe,
        operacion(cafe, 'erosion'),
        operacion(cafe, 'dilatacion'),
        operacion(cafe, 'apertura'),
        operacion(cafe, 'cierre')]
title = ["Original","Erosion","Dilatacion","Apertura","Cierre"]
create_subplots(data, title)
```

# PUNTO #3
Crear un código que pueda separar los objetos de la imagen por tamaños, permitiendo al usario elegir el tamaño. Para los resutlados mostrar en una sola gráfica dos figuras, una con la imagen original y la otra con el elemento que se eliminó o separó.

```python
circulos = mpimg.imread('ImageProcessing/Parcial1/balls.jpg')
circulos_gray = cv2.cvtColor(circulos, cv2.COLOR_BGR2GRAY) #Se transforma a escala de grises

#Entrada del usuario
radio = int(input("Ingrese el radio de los círculos a buscar: "))
tolerancia = int(input("Ingrese la tolerancia en el radio de los círculos a buscar: "))

#Detección de círculos por Transformada de Hough
circles = cv2.HoughCircles(circulos_gray, cv2.HOUGH_GRADIENT, dp=1, minDist=20, param1=50,
                           param2=30, minRadius=radio-tolerancia, maxRadius=radio+tolerancia)

#Crear una imagen copia en negro para sobreescribir los circulos selectos
circulos_selectos = np.zeros_like(circulos_gray)

detected_circles = []
if circles is not None:
  # Conversión de círculos a escalares
  circles = np.uint16(np.around(circles))

  # Detección e impresión de los círculos selectos
  for circle in circles[0, :]:
    x, y, radius = circle
    detected_circles.append((x, y, radius))
    cv2.circle(circulos_selectos, (x, y), radius, (255, 255, 255), thickness=-1)

data = [circulos, circulos_selectos]
title = ["Original",f"Circulos Detectados en radio {radio} y tolerancia {tolerancia}"]
create_subplots(data, title)
```
