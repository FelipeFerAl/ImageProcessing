```python
def plot_image(image, title):
  """
  Esta función recibe una imagen y la muestra en pantalla ajustando
  el tamaño en base a las dimensiones del array de la imagen.
  Entrada: Imagen como array
  Salida: Imagen en pantalla
  """
  # plt.figure(figsize=(image.shape[0]*2/100, image.shape[1]*2/100))
  plt.imshow(image)
  plt.title(title)
  plt.axis('off')
  plt.show()
```

```python
def create_subplots(data_list, titles):
  """
  Esta función recibe un conjunto de datos (imágenes) para imprimir en
  pantalla junto con sus respectivos títulos.
  Entrada: Lista de imágenes y lista de títulos
  Salida: Imágenes en pantalla
  """
  plt.figure(figsize=(16, 12))
  num_images = len(data_list)
  num_cols = 3
  num_rows = (num_images + num_cols - 1) // num_cols
  for i in range(num_images):
    plt.subplot(num_rows, num_cols, i + 1)
    plt.imshow(data_list[i])
    plt.title(titles[i])
    plt.axis('off')  # Oculta los ejes
  plt.show()
```

Construir una función debidamente documentada, que reciba un parámetro llamado "color", con el cual se indique los colores de caramelos que se quieren mostrar en la siguiente imagen:

```python
#@title Función
class color_ID:
  def __init__(self, imagen):
    """
    Inicializa una clase para guardar la
    imagen con que se quiere trabajar
    Entrada: Imagen a procesar
    """
    self.imagen = imagen
    self.hsv = cv2.cvtColor(imagen, cv2.COLOR_RGB2HSV)  #Conversión a HSV
    self.kernel = np.ones((5,5), np.uint8)  #Kernel para pasar por las imágenes

  def amarillo(self):
    """
    Identificación del color amarillo
    Entrada: None
    Salida: Imagen con objetos de color amarillo
    """
    #Creación de las máscaras
    low_yellow = np.array([20,100,100])
    upper_yellow = np.array([30,255,255])
    #Aplicación de la máscara
    mask = cv2.inRange(self.hsv, low_yellow, upper_yellow)
    res = cv2.bitwise_and(self.imagen, self.imagen, mask=mask)
    # Aplica una operación de opening para eliminar el ruido tipo salt
    op = cv2.morphologyEx(res, cv2.MORPH_OPEN, self.kernel)
    # Aplica una operación de closing después de la opening para eliminar tanto el ruido tipo salt como el ruido tipo pepper
    cl = cv2.morphologyEx(op, cv2.MORPH_CLOSE, self.kernel)
    return cl

  def verde(self):
    """
    Identificación del color verde
    Entrada: None
    Salida: Imagen con objetos de color verde
    """
    #Creación de las máscaras
    low_green = np.array([40,100,100])
    upper_green = np.array([80,255,255])
    #Aplicación de la máscara
    mask = cv2.inRange(self.hsv, low_green, upper_green)
    res = cv2.bitwise_and(self.imagen, self.imagen, mask=mask)
    # Aplica una operación de opening para eliminar el ruido tipo salt
    op = cv2.morphologyEx(res, cv2.MORPH_OPEN, self.kernel)
    # Aplica una operación de closing después de la opening para eliminar tanto el ruido tipo salt como el ruido tipo pepper
    cl = cv2.morphologyEx(op, cv2.MORPH_CLOSE, self.kernel)
    return cl

  def azul(self):
    """
    Identificación del color azul
    Entrada: None
    Salida: Imagen con objetos de color azul
    """
    #Creación de las máscaras
    low_blue = np.array([100,100,100])
    upper_blue = np.array([140,255,255])
    #Aplicación de la máscara
    mask = cv2.inRange(self.hsv, low_blue, upper_blue)
    res = cv2.bitwise_and(self.imagen, self.imagen, mask=mask)
    # Aplica una operación de opening para eliminar el ruido tipo salt
    op = cv2.morphologyEx(res, cv2.MORPH_OPEN, self.kernel)
    # Aplica una operación de closing después de la opening para eliminar tanto el ruido tipo salt como el ruido tipo pepper
    cl = cv2.morphologyEx(op, cv2.MORPH_CLOSE, self.kernel)
    return cl

  def purpura(self):
    """
    Identificación del color azul
    Entrada: None
    Salida: Imagen con objetos de color azul
    """
    #Creación de las máscaras
    low_purple = np.array([120,50,50])
    upper_purple = np.array([160,255,255])
    #Aplicación de la máscara
    mask = cv2.inRange(self.hsv, low_purple, upper_purple)
    res = cv2.bitwise_and(self.imagen, self.imagen, mask=mask)
    # Aplica una operación de opening para eliminar el ruido tipo salt
    op = cv2.morphologyEx(res, cv2.MORPH_OPEN, self.kernel)
    # Aplica una operación de closing después de la opening para eliminar tanto el ruido tipo salt como el ruido tipo pepper
    cl = cv2.morphologyEx(op, cv2.MORPH_CLOSE, self.kernel)
    return cl

  def rosado(self):
    """
    Identificación del color rosado
    Entrada: None
    Salida: Imagen con objetos de color rosado
    """
    #Creación de las máscaras
    low_pink = np.array([140,50,50])
    upper_pink = np.array([170,200,255])
    #Aplicación de la máscara
    mask = cv2.inRange(self.hsv, low_pink, upper_pink)
    res = cv2.bitwise_and(self.imagen, self.imagen, mask=mask)
    # Crea un kernel para operaciones morfológicas
    kernel = np.ones((6,6), np.uint8)
    # Aplica una operación de opening para eliminar el ruido tipo salt
    op = cv2.morphologyEx(res, cv2.MORPH_OPEN, kernel)
    # Aplica una operación de closing después de la opening para eliminar tanto el ruido
    cl = cv2.morphologyEx(op, cv2.MORPH_CLOSE, kernel)
    return cl

  def rojo(self):
    """
    Identificación del color naranja
    Entrada: None
    Salida: Imagen con objetos de color naranja
    """
    # Creación de la máscara
    low_red = np.array([170,50,50])
    upper_red = np.array([180,255,255])
    #Aplicación de la máscara
    mask = cv2.inRange(self.hsv, low_red, upper_red)
    res = cv2.bitwise_and(self.imagen, self.imagen, mask=mask)
    # Aplica una operación de opening para eliminar el ruido tipo salt
    op = cv2.morphologyEx(res, cv2.MORPH_OPEN, self.kernel)
    # Aplica una operación de closing después de la opening para eliminar tanto el ruido
    cl = cv2.morphologyEx(op, cv2.MORPH_CLOSE, self.kernel)
    return cl


  def naranja(self):
    """
    Identificación del color naranja
    Entrada: None
    Salida: Imagen con objetos de color naranja
    """
    # Creación de la máscara
    low_orange = np.array([0,100,100])
    upper_orange = np.array([10,255,255])
    #Aplicación de la máscara
    mask = cv2.inRange(self.hsv, low_orange, upper_orange)
    res = cv2.bitwise_and(self.imagen, self.imagen, mask=mask)
    # Aplica una operación de opening para eliminar el ruido tipo salt
    op = cv2.morphologyEx(res, cv2.MORPH_OPEN, self.kernel)
    # Aplica una operación de closing después de la opening para eliminar tanto el ruido
    cl = cv2.morphologyEx(op, cv2.MORPH_CLOSE, self.kernel)
    return cl

  def negro(self):
    """
    Identificación del color negro
    Entrada: None
    Salida: Imagen con objetos de color negro
    """
    # Creación de la máscara
    low_black = np.array([0,0,0])
    upper_black = np.array([180,255,50])
    #Aplicación de la máscara
    mask = cv2.inRange(self.hsv, low_black, upper_black)
    res = cv2.bitwise_and(self.imagen, self.imagen, mask=mask)
    # Aplica una operación de opening para eliminar el ruido tipo salt
    op = cv2.morphologyEx(res, cv2.MORPH_OPEN, self.kernel)
    # Aplica una operación de closing después de la opening para eliminar tanto el ruido
    cl = cv2.morphologyEx(op, cv2.MORPH_CLOSE, self.kernel)
    return cl

  def cafe(self):
    """
    Identificación del color cafe
    Entrada: None
    Salida: Imagen con objetos de color cafe
    """
    # Creación de la máscara
    low_brown = np.array([0,20,0])
    upper_brown = np.array([30,100,50])
    #Aplicación de la máscara
    mask = cv2.inRange(self.hsv, low_brown, upper_brown)
    res = cv2.bitwise_and(self.imagen, self.imagen, mask=mask)
    # Aplica una operación de opening para eliminar el ruido tipo salt
    op = cv2.morphologyEx(res, cv2.MORPH_OPEN, self.kernel)
    # Aplica una operación de closing después de la opening para eliminar tanto el ruido
    cl = cv2.morphologyEx(op, cv2.MORPH_CLOSE, self.kernel)
    return cl
```
